## Canonical covariance analysis

### Set up and get data

```python
## Import modules
import os
import abct
import numpy as np
from scipy import io
import plotly.express as px
import plotly.subplots as sp

## Unpack data
data = io.loadmat(os.path.join(".", "data", "hcp_data.mat"))
connectivity = data["connectivity"]
activity = data["activity"]
coactivity = activity.T @ activity

k = 5           # Number of components 
repl = 100      # Number of replicates
```

### Co-activity gradients

```python
# Weighted gradients
V = abct.gradients(coactivity, k)

# Binary gradients
V_bin = abct.gradients(coactivity, k, "binary")
```

### Residual degree

```python
d_ = abct.degrees(coactivity, "residual")
```
